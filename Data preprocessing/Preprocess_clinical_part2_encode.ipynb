```python
""" Step 1: Data Preparation """
# 1.1.Import the libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

from sklearn.preprocessing import OrdinalEncoder

sns.set_style("whitegrid")
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load data
from google.colab import drive
drive.mount('/content/drive')


data = pd.read_csv('/content/drive/MyDrive/Data/Survival(Pre-Processed).csv')
```

```python
data['OS.time'].describe()
```

```python
# Split by median
data['Risk'] = np.where(data['OS.time'] >= 839, 0, 1)
```

```python
data.info()
```

```python
# Drop not relevant variables
data = data.drop(['gender','type', 'OS'], axis=1)
```

```python
data.head(5)
```

```python
# Encode catergorical

df_scale = data.drop(['bcr_patient_barcode','OS.time','age_at_initial_pathologic_diagnosis','initial_pathologic_dx_year'], axis=1)

df_scale1 = OrdinalEncoder().fit_transform(df_scale)
col = list(df_scale.columns)
df_scale2 = pd.DataFrame(df_scale1,columns=col)
```

```python
df =pd.concat([data[['bcr_patient_barcode','OS.time','age_at_initial_pathologic_diagnosis','initial_pathologic_dx_year']],df_scale2],axis=1)
df.head(3)
```
